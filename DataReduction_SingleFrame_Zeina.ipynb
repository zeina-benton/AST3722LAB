```python
import numpy as np
import pylab as pl
pl.rcParams['image.origin'] = 'lower' # make images display right-side-up
pl.style.use('dark_background')
```

```python
import glob
import os
os.chdir('/Users/zeinakellybenton/obs_tech/M44-lab3')
```

```python
#M44 Bee hive Cluster
folder_path= '/Users/zeinakellybenton/obs_tech'
subfolders = [f.path for f in os.scandir(folder_path) if f.is_dir()]

darks_lab3_files = []
bias_lab3_files = []
flats_lab3_files = []

for subfolder in subfolders:
    file_list = os.listdir(subfolder)
    
    if 'darks-lab3' in subfolder:
        darks_lab3_files = file_list
    elif 'bias-lab3' in subfolder:
        bias_lab3_files = file_list
    elif 'flat-lab3' in subfolder:
        flats_lab3_files = file_list

print("Files in 'darks-lab3':", darks_lab3_files)
```

```python
print("Files in 'bias-lab3':", bias_lab3_files)
```

```python
print("Files in 'flat-lab3':", flats_lab3_files)
```

```python
darks_10s = [file for file in darks_lab3_files if '10s' in file]

print("Darks with 10s exposure:", darks_10s)
```

```python
from astropy.io import fits

folder_path = '/Users/zeinakellybenton/obs_tech/bias-lab3' 
bias_data = []

for file in bias_lab3_files:
    full_file_path = os.path.join(folder_path, file)
    
    with fits.open(full_file_path) as hdul:
        data = hdul[0].data  
        bias_data.append(data)

bias_data = np.array(bias_data)
```

```python
mean_bias = np.mean(bias_data, axis=0)
stddev_bias = np.std(bias_data, axis=0)
readnoise = np.mean(stddev_bias)
```

```python
mean_bias_noise = readnoise / np.sqrt(len(bias_lab3_files))
mean_bias_noise
```

```python
#calling only the clear

folder_path = '/Users/zeinakellybenton/obs_tech/flat-lab3' 
flats_data = []

for file in flats_lab3_files:
    if 'Flat_darks_0.2sec_Bin2_filter-Clear' in file:
        full_file_path = os.path.join(folder_path, file)
        
        with fits.open(full_file_path) as hdul:
            data = hdul[0].data  
            flats_data.append(data)


flats_data = np.array(flats_data)
```

```python
mean_clear_flat = np.mean(flats_data, axis=0)
bias_subtracted_clear_flat = mean_clear_flat - mean_bias
```

```python
#my data (with avery and carli) were set to gain to be 0 when doing these so we have to skip
#all of the multiply data by gain and tp ignore it (spoke with Tayar in class)
```

```python
flat_poisson_uncertainty = (bias_subtracted_clear_flat)**0.5
```

```python
flat_singleframe_uncertainty = (flat_poisson_uncertainty**2 + readnoise**2)**0.5
```

```python
flat_total_uncertainty = ((flat_singleframe_uncertainty/np.sqrt(len(flats_data)))**2 + mean_bias_noise**2)**0.5
```

```python
mean_of_flat = np.mean(bias_subtracted_clear_flat)
mean_of_flat
```

```python
normalized_mean_clear_flat = bias_subtracted_clear_flat / mean_of_flat
np.mean(normalized_mean_clear_flat)
```

```python
pl.figure(figsize=(12,8))

_=pl.imshow(normalized_mean_clear_flat, origin='lower', interpolation='none')
_=pl.colorbar()
_=pl.title("Normalized Flat")
_=pl.xlabel("X pixel")
_=pl.ylabel("Y pixel")
```

```python
normalized_flat_uncertainty = flat_total_uncertainty / mean_of_flat
```

```python
pl.xlabel("Uncertainty (Std. Deviation)")
pl.ylabel("Number of pixels")
_=pl.hist(normalized_flat_uncertainty.ravel(), bins=50, log=True)
```

```python
pl.figure(figsize=(12,8))
pl.title("Signal-to-noise ratio of flat field")
pl.imshow(normalized_mean_clear_flat / normalized_flat_uncertainty, origin='lower', interpolation='none')
pl.colorbar()
_=pl.xlabel("X pixel")
_=pl.ylabel("Y pixel")
```

```python
folder_path = '/Users/zeinakellybenton/obs_tech/darks-lab3'
darks_10s_data = []

for file in darks_10s:
    full_file_path = os.path.join(folder_path, file)
    
    with fits.open(full_file_path) as hdul:
        data = hdul[0].data 
        darks_10s_data.append(data)

darks_10s_data = np.array(darks_10s_data)
```

```python
average_10s_dark = np.mean(darks_10s_data, axis=0)
stddev_10s_dark = np.std(darks_10s_data, axis=0)
```

```python
vmax = np.max(average_10s_dark)
#had to avoid saturation since gain is 0

pl.figure(figsize=(12, 8))
pl.imshow(average_10s_dark, origin='lower', interpolation='none', vmax=vmax)
pl.colorbar()
_=pl.title("Average Dark")
_=pl.xlabel("X pixel")
_=pl.ylabel("Y pixel")
```

```python
uncertainty_on_10s_mean_dark = stddev_10s_dark / np.sqrt(len(darks_10s_data))
```

```python
folder_path = '/Users/zeinakellybenton/obs_tech/M44-lab3'

all_files = os.listdir(folder_path)

science_lab3_files = [file for file in all_files if 'Light_darks_10sec_Bin2_filter-Clear' in file]


science_data = []

for file in science_lab3_files:
    full_file_path = os.path.join(folder_path, file)
    
    with fits.open(full_file_path) as hdul:
        data = hdul[0].data
        science_data.append(data)

science_data = np.array(science_data)
```

```python
image_to_display = science_data[0] #2D

pl.figure(figsize=(16,5))
im = pl.subplot(1,2,1).imshow(image_to_display, origin='lower', interpolation='none', vmax=2500,)
_=pl.title("M44 Clear (no-filter) image")
_=pl.xlabel("X pixel")
_=pl.ylabel("Y pixel")
pl.colorbar(im)

_ = pl.subplot(1,2,2).hist(science_data.ravel(), bins=50, log=True)
_ = pl.xlabel("Counts")
_ = pl.ylabel("Number of pixels")
```

```python
m44_clear_darksub = science_data.squeeze() - average_10s_dark
```

```python
sum_inside_sqrt = m44_clear_darksub + readnoise**2 + uncertainty_on_10s_mean_dark**2
sum_inside_sqrt = np.maximum(sum_inside_sqrt, 0)  # Set negative values to 0
m44_clear_darksub_uncertainty = np.sqrt(sum_inside_sqrt)
```

```python
pl.figure(figsize=(16,5))
im = pl.subplot(1,2,1).imshow(m44_clear_darksub, origin='lower', interpolation='none', vmax=600, vmin=200)
_=pl.title("M44 Clear, dark-subtracted image")
_=pl.xlabel("X pixel")
_=pl.ylabel("Y pixel")
pl.colorbar(im)

_ = pl.subplot(1,2,2).hist(m44_clear_darksub.ravel(), bins=50, log=True)
_ = pl.xlabel("Counts")
_ = pl.ylabel("Number of pixels")
```

```python
m44_clear_darksub_flattened = m44_clear_darksub / normalized_mean_clear_flat
```

```python
inside_sqrt = m44_clear_darksub + readnoise**2 + uncertainty_on_10s_mean_dark**2
inside_sqrt = np.maximum(inside_sqrt, 0)  # Ensure no negative values

m44_clear_darksub_uncertainty = np.sqrt(inside_sqrt)
```

```python
m44_clear_darksub_flattened_uncertainty = ((
    m44_clear_darksub_uncertainty**2 / normalized_mean_clear_flat**2 +
    normalized_flat_uncertainty**2 * m44_clear_darksub**2 / normalized_mean_clear_flat**4
))**0.5
```

```python
pl.figure(figsize=(16,5))
im = pl.subplot(1,2,1).imshow(m44_clear_darksub_flattened, origin='lower', interpolation='none', vmax=1000, vmin=200)
pl.colorbar(im)
_=pl.title("M44 Clear, dark-subtracted, flattened image")
_=pl.xlabel("X pixel")
_=pl.ylabel("Y pixel")

_ = pl.subplot(1,2,2).hist(m44_clear_darksub_flattened.ravel(), bins=50, log=True)
_ = pl.xlabel("Counts")
_ = pl.ylabel("Number of pixels")
```

```python
pl.figure(figsize=(16,5))
SNR = m44_clear_darksub_flattened / m44_clear_darksub_flattened_uncertainty
im = pl.subplot(1,2,1).imshow(SNR, origin='lower', interpolation='none', vmin=0, vmax=30)
_=pl.title("M44 SNR")
_=pl.xlabel("X pixel")
_=pl.ylabel("Y pixel")
pl.colorbar(im)

_ = pl.subplot(1,2,2).hist(SNR.ravel(), bins=50, log=True)
_ = pl.xlabel("SNR")
_ = pl.ylabel("Number of pixels")
```
